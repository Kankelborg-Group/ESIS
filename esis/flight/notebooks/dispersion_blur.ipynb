```python
%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
from kgpy.img.coalignment import image_coalignment as img_align

plt.rcParams['figure.figsize'] = [10, 10]
```

Quick and dirty production of a convolution kernel used to simulate blur due to ESIS type dispersion

```python
kernel = np.zeros((11,11))
origin = (5,5)
width_x,width_y= (.5,8)
x,y = img_align.get_img_coords(kernel)
x -= origin[0]
y -= origin[1]
gaussian = np.exp(-(np.square(x)/width_x+np.square(y)/width_y))

fig,ax = plt.subplots()
ax.imshow(gaussian)
```

```python
angles = -(np.arange(4))*45 - 22.5 -90
angles = np.deg2rad(angles)

gaussians = []
for angle in angles:
    x_prime = x*np.cos(angle) + y*np.sin(angle)
    y_prime = x*np.sin(angle) - y*np.cos(angle)
    gaussians.append(np.exp(-(np.square(x_prime)/width_x+np.square(y_prime)/width_y)))

z = 0
fig,ax = plt.subplots(2,2)

for i in range(2):
    for j in range(2):
        ax[i,j].imshow(gaussians[z])
        z+=1






```
